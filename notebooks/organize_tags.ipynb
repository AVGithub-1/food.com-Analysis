```python
import numpy as numpy
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('tags.csv')

total_tags = set([])
```

```python
def get_all_tags(tag_list):
    tag_list = tag_list[1:-1]
    tag_list = tag_list.replace("'", "")
    tag_list  = tag_list.replace(" ", "")
    tag_list = list(tag_list.split(','))
    total_tags.update(tag_list)

    return tag_list
```

```python
df['tags'] = df['tags'].apply(get_all_tags)
```

```python
import csv

# Convert set to list
my_list = list(total_tags)

# Specify the file name
file_name = 'total_tags.csv'

# Write list to CSV file
with open(file_name, 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['tagName,'])
    writer.writerows(map(lambda x: [x+","], my_list))
    
print(f"CSV file '{file_name}' has been created successfully.")
```
